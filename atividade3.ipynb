5. Apresente e depois elimine os outliers do campo salário. Considere que um outlier será dado pelo valor da média +- 2 vezes o seu desvio padrão.

```python
sal_mean = df["Salary"].mean()
std_sal = df["Salary"].std()

limite_inf = sal_mean - 2 * std_sal
limite_sup = sal_mean + 2 * std_sal

print(f'Mean: {sal_mean}')
print(f'Std: {std_sal}')
print(f"Interval: [{limite_inf:.2f}, {limite_sup:.2f}]")

outliers = df[(df["Salary"] < limite_inf) | (df["Salary"] > limite_sup)]
outliers
```

```python
df = df[(df["Salary"] >= limite_inf) & (df["Salary"] <= limite_sup)]
df
```

6. Verifique se há alguma dado inconsistente considerando os campos idade e anos de experiência. Considere que as pessoas só podem começar a trabalhar após os 16 anos de idade.

```python
inconsistencies = df[df["Age"] - df["Years of Experience"] < 16]
inconsistencies
```

```python
df = df[~(df["Age"] - df["Years of Experience"] < 16)]
df
```

7.  Analise os dados e aplique mais alguma melhoria ao dataset resultante das transformações anteriores.

```python
df2 = df.copy()
df2["Job Title"] = df2["Job Title"].str.strip().str.title()
df2
```

```python
df2.nunique()
```

```python
jobs = []
for job in df2['Job Title']:
    if job not in jobs:
        jobs.append(job)
sorted(jobs)
```

```python
education_levels = []
for e in df2['Education Level']:
    if e not in education_levels:
        education_levels.append(e)
sorted(education_levels)
```
