Importação das bibliotecas necessárias para os exemplos

```python
import matplotlib.pyplot as plt
import numpy as np
```

Criação de dois vetores que representam as coordenadas dos eixos x e y respectivamente.

```python
#Cria um vetor com os elementos de 0 a 9
X = np.arange(10)
#Cria outro vetor que recebe o dobro dos valores de x
Y = 2 * X

print(X)
print(Y)
```

A criação de um gráfico de linha é feita pela função **plot**. O código abaixo cria e apresenta um gráfico de linha padrão com os valores dos vetores X e Y

```python
plt.plot(X, Y)
plt.show()
```

O parâmetro fmt configura o gráfico para que ele apresente marcadores quadrados (s), ligados por uma linha tracejada (--) em preto (k).

Outros valores para o atributo fmt podem ser encontrados em:
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
plt.plot(X, Y,'s--k')
plt.show()
```

O código abaixo cria um vetor com 31 posições e em cada uma delas armazena os números 1 a 31 em sequência para representar os dias do mês e outro vetor, também com 31 posições, com valores gerados aleatóriamente entre 10 e 30 para representar a temperatura máxima de cada dia.

```python
diaMes = np.arange(1,32)
tempMax = np.random.randint(10,30, size=31)
print(diaMes)
print(tempMax)
```

Para adicionar título ao gráfico e aos eixos x e y, deve-se usar, respectivamente, as funções **title**, **xlabel** e **ylabel**.

```python
plt.plot(diaMes, tempMax,'o-')
plt.title('Temperatura máxima diária em julho de 2023')
plt.xlabel('Dia')
plt.ylabel('Temperatura')
plt.show()
```

O código a seguir cria um novo vetor, com a temperatura mínima do dia que é dada por sua temperatura máxima menos uma diferença aleatória entre 5 e 15 graus.

```python
diferenca = np.random.randint(5,15, size=31)
tempMin = tempMax - diferenca

print(tempMin)
```

Para adicionar duas ou mais séries em um gráfico deve-se indicar no parâmetro **label** o texto de identificação que aparecerá na legenda. Além disto, deve-se chamar a função legend que mostra efetivamente a legenda.

```python
plt.plot(diaMes, tempMax, 'o-', label='Máxima')
plt.plot(diaMes, tempMin, 'o-r', label='Mínima')

plt.title('Temperatura máxima diária em julho de 2023')
plt.xlabel('Dia')
plt.ylabel('Temperatura')
plt.legend()
plt.show()
```

A criação de gráficos de dispersão pode ser feita usando a função **scatter**, os parâmetros e demais funções usadas com a função **plot** também podem ser usados com a função **scatter**.

```python
plt.scatter(tempMin, tempMax)
plt.title('Relação entre as temperaturas máximas e mínimas diárias')
plt.xlabel('Temperatura Mínima')
plt.ylabel('Temperatura Máxima')
plt.show()
```

A função **scatter** também pode ser usada para criar um gráfico de bolhas. Além dos valores constantes nos eixos x e y, uma terceira informação é representada pelo tamanho das bolhas.

O gráfico abaixo apresenta as cidades mais populosas de Santa Catarina no eixo x, sua população no eixo y e o valor do seu Produto Interno Bruto (PIB) no tamanho das bolhas, por meio do parâmetro **s**. O parâmetro **c** pode ser usado para definir as cores de cada bolha (neste exemplo foram definidas aleatoriamente). Caso ele não seja definido, todas as bolhas terão a mesma cor.

Os limites dos valores dos eixos x e y podem ser alterados por meio das funções **xlim** e **ylim**.

```python
cidades = ['Joinville','Florianópolis','Blumenau','São José','Itajaí']
populacao = [616323,537213,361261,270295,264054]
pib=[1360.4,910.3,770.8,640.5,1130.1]

cores = np.random.rand(5)
#plt.scatter(cidades, populacao)
plt.scatter(cidades, populacao,s=pib, c=cores)
plt.title('População e PIB das cidades mais populosas de Santa Catarina')
plt.xlabel('Cidades')
plt.ylabel('População')
plt.ylim(200000, 700000)
plt.show()
```

A criação de gráficos de pizza é feita pela função **pie**, que recebe primeiramente como parâmetro os valores númericos que determinam o "tamanho" da fatia de cada categoria. O parâmetro **labels** recebe uma lista com os textos relacionados a cada valor e o parâmetro **autopct** apresenta o percentual referente a cada categoria. Caso este parâmetro não seja passado nenhum valor numérico é mostrado no gráfico.

```python
cidades = ['Joinville','Florianópolis','Blumenau','São José','Itajaí']
populacao = [616323,537213,361261,270295,264054]

plt.pie(populacao,labels=cidades,autopct="%.2f%%")
plt.title('Cidades mais populosas de Santa Catarina')
plt.show()
```

Para apresentar o valor absoluto e não o percentual usando o parâmetro autopct deve-se criar uma função que calcula este valor a partir do percentual.

```python
plt.pie(populacao,labels=cidades,autopct=lambda x:f"{sum(populacao)*x/100:.0f}")
plt.title('Cidades mais populosas de Santa Catarina')
plt.show()
```

Para criar gráficos de colunas e barras deve-se usar, respectivamente, as funções **bar** e **barh**.

```python
plt.bar(cidades,populacao)
plt.title('Cidades mais populosas de Santa Catarina')
plt.xlabel('Cidade')
plt.ylabel('Quantidade de habitantes')
plt.show()
```

```python
plt.barh(cidades,populacao)
plt.title('Cidades mais populosas de Santa Catarina')
plt.xlabel('Cidade')
plt.ylabel('Quantidade de habitantes')
plt.show()
```

Para criar um histograma de frequência deve-se usar a função **hist**. Um parâmetro essencial desta função é o **bins** que indica a quantidade de classes ou os limites de cada uma das classes.
O atributo rwidth define se deve deixar espaço entre as colunas do histograma ou não. Se o parâmetor receber o valor 1, ou simplesmente não for definido, as colunas ficarão juntas umas das outras. Quanto menor o valor maior o espaçamento.

Histograma de classes que distribui as temperaturas máximas em 5 classes diferentes.

```python
plt.hist(tempMax,bins=5, rwidth=0.99)
plt.title('Distribuição de frequência da Temperatura Máxima')
plt.xlabel('Temperatura')
plt.ylabel('Frequência')
plt.show()
```

O parâmetro bins pode receber uma lista com os limites de cada uma das classes.

```python
plt.hist(tempMax,bins=[10,15,20,25,30], rwidth=0.99)
plt.title('Distribuição de frequência da Temperatura Máxima')
plt.xlabel('Temperatura')
plt.ylabel('Frequência')
plt.show()
```
